#### clean planet data

replace 230429, 230623 with the mean in planet

```python
for idx, row in df_planet.iterrows():
    mean_0429 = (row["230426"] + row["230430"]) / 2
    mean_0623 = (row["230611"] + row["230626"]) / 2
    df_planet.loc[idx, "230429"] = mean_0429
    df_planet.loc[idx, "230623"] = mean_0623
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(df_planet.values[:,1:].T, alpha=0.3);
ax.set_xticks(
    range(len(df_planet.columns[1:])),
    df_planet.columns[1:],
    rotation=45,
);
ax.set_title('Planet after replacing with the mean');
```

```python
from datetime import datetime

date_format = "%Y%m%d"

p4m_columns = df_p4m.columns[1:].values.tolist()
planet_columns = df_planet.columns[1:].values.tolist()

def calculate_delta_days(day1, day2):
    day1 = "20" + day1
    day2 = "20" + day2

    day1 = datetime.strptime(day1, date_format)
    day2 = datetime.strptime(day2, date_format)

    delta = (day2 - day1).days
    return delta

p4m_indexes = [ 
    calculate_delta_days("230320", day) for day in p4m_columns
]

planet_indexes = [ 
    calculate_delta_days("230324", day) for day in planet_columns
]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharey=True, figsize=(12, 12))

for idx, row in df_p4m.iterrows():
    axes[0].plot(p4m_indexes, row.values[1:], alpha=0.5)
axes[0].set_title("P4M", fontsize=15)
axes[0].set_xlim([0, 130]);

for idx, row in df_planet.iterrows():
    axes[1].plot(planet_indexes, row.values[1:], alpha=0.5)
axes[1].set_title("Planet", fontsize=15)
axes[1].set_xlim([0, 130]);


# axes[0].plot(p4m_indexes, p4m_max, color="yellow", linewidth=5, linestyle="--");
# axes[0].plot(p4m_indexes, p4m_min, color="yellow", linewidth=5, linestyle="--");

# axes[1].plot(planet_indexes, planet_max, color="orange", linewidth=5, linestyle="--");
# axes[1].plot(planet_indexes, planet_min, color="orange", linewidth=5, linestyle="--");
```

#### Generate Heatmap

```python
# load yield data
yield_data = pd.read_csv(os.path.join(DATA_DIR, "Yield", "yield_result.csv"))
```

```python
yield_data.info()
```

```python
df_p4m = df_p4m[df_p4m["fid"].isin(yield_data['fid_1'])]
df_planet = df_planet[df_planet["fid"].isin(yield_data["fid_1"])]
```

```python
heatmap_p4m_data = df_p4m.copy()
heatmap_p4m_data["yield"] = yield_data["final_yield"]
heatmap_p4m_data.drop(columns=["fid"], inplace=True)
```

```python
heatmap_planet_data = df_planet.copy()
heatmap_planet_data["yield"] = yield_data["final_yield"]
heatmap_planet_data.drop(columns=["fid"], inplace=True)
```

```python
p4m_corr_matrix = heatmap_p4m_data.corr()
planet_corr_matrix = heatmap_planet_data.corr()
```

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.heatmap(
    p4m_corr_matrix["yield"].values.reshape(1, -1),
    cmap="viridis"
);

ax.set_xticks(
    range(len(p4m_corr_matrix.columns)),
    p4m_corr_matrix,
    rotation=45,
);

ax.set_yticks([]);
ax.set_ylabel("yield");
ax.set_title("P4M vs Yield HeatMap");
```

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.heatmap(
    planet_corr_matrix["yield"].values.reshape(1, -1),
    cmap="viridis"
);

ax.set_xticks(
    range(len(planet_corr_matrix.columns)),
    planet_corr_matrix,
    rotation=45,
);

ax.set_yticks([]);
ax.set_ylabel("yield");
ax.set_title("Planet vs Yield HeatMap");
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.histplot(data=yield_data["final_yield"]);
ax.set_xlabel("yield (lbs)");

yield_mean = np.mean(yield_data["final_yield"])
yield_std = np.std(yield_data["final_yield"])

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

ax.text(
    10, 250, f"mean: {yield_mean:.3f}\nstd: {yield_std:.3f}",
    fontsize=20,
    bbox=props,

);
```
